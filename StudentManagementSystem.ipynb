```python
import pandas as pd
import os
```

```python
FILE_NAME = "students.csv"

if not os.path.exists(FILE_NAME):
    df = pd.DataFrame(columns=[
        "ID",
        "Name",
        "Age",
        "Gender",
        "Course",
        "Marks"
    ])
    df.to_csv(FILE_NAME, index=False)

print("Database Ready")
```

```python
def add_student():
    df = pd.read_csv(FILE_NAME)

    sid = input("Student ID: ")
    name = input("Name: ")
    age = int(input("Age: "))
    gender = input("Gender: ")
    course = input("Course: ")
    marks = float(input("Marks: "))

    new_data = {
        "ID": sid,
        "Name": name,
        "Age": age,
        "Gender": gender,
        "Course": course,
        "Marks": marks
    }

    df.loc[len(df)] = new_data
    df.to_csv(FILE_NAME, index=False)

    print("Student Added Successfully")
```

```python
def view_students():
    df = pd.read_csv(FILE_NAME)
    display(df)
```

```python
def search_student():
    df = pd.read_csv(FILE_NAME)

    sid = input("Enter Student ID: ")

    result = df[df["ID"] == sid]

    if len(result) > 0:
        display(result)
    else:
        print("Student Not Found")
```

```python
def update_student():
    df = pd.read_csv(FILE_NAME)

    sid = input("Enter Student ID: ")

    if sid in df["ID"].values:
        name = input("New Name: ")
        course = input("New Course: ")
        marks = float(input("New Marks: "))

        df.loc[df["ID"] == sid, "Name"] = name
        df.loc[df["ID"] == sid, "Course"] = course
        df.loc[df["ID"] == sid, "Marks"] = marks

        df.to_csv(FILE_NAME, index=False)
        print("Updated Successfully")
    else:
        print("Student Not Found")
```

```python
def delete_student():
    df = pd.read_csv(FILE_NAME)

    sid = input("Enter Student ID: ")

    df = df[df["ID"] != sid]

    df.to_csv(FILE_NAME, index=False)

    print("Deleted Successfully")
```

```python
def calculate_grade(mark):

    if mark >= 90:
        return "A+"
    elif mark >= 80:
        return "A"
    elif mark >= 70:
        return "B"
    elif mark >= 60:
        return "C"
    elif mark >= 50:
        return "D"
    else:
        return "F"
```

```python
df = pd.read_csv(FILE_NAME)

df["Grade"] = df["Marks"].apply(calculate_grade)

display(df)
```

```python
while True:

    print("\n===== Student Management System =====")

    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Show Grades")
    print("7. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        view_students()

    elif choice == "3":
        search_student()

    elif choice == "4":
        update_student()

    elif choice == "5":
        delete_student()

    elif choice == "6":
        df = pd.read_csv(FILE_NAME)
        df["Grade"] = df["Marks"].apply(calculate_grade)
        display(df)

    elif choice == "7":
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
